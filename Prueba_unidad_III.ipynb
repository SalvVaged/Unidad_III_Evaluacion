**Ejemplo:**

### Descripci√≥n

- Ingresos **‚â§ $10,000**: no pagan impuestos.  
- Ingresos **entre $10,001 y $50,000**: pagan **10% sobre el exceso** de $10,000.  
- Ingresos **> $50,000**: pagan **20% sobre el exceso de $50,000** m√°s **$4,000**.  
- Mayores de **65 a√±os**: reducci√≥n del **5% en el impuesto total**.  
- Por cada **dependiente**, se reduce **$1,000 del ingreso gravable** antes de calcular el impuesto.  

---

### ‚úÖ Salida esperada
El programa debe calcular y mostrar el impuesto a pagar.  

**Ejemplo:**
- Entrada: `ingreso=60000, edad=70, dependientes=2`  
- Salida: `Impuesto: 5320`  

üìå *Explicaci√≥n:*  
- Ingreso gravable = 60000 - (1000 √ó 2) = **58000**  
- Impuesto base = 4000 + 20% de (58000 - 50000) = **5600**  
- Descuento por edad = 5% de 5600 = **280**  
- Impuesto final = 5600 - 280 = **5320**

```python
ingr = int(input("Ingrese sus ingresos anuales: "))
edad = int(input("Ingrese su edad: "))
dep = int(input("Ingrese el numero de dependientes: "))

if ingr < 10000:
    print("No pagas impuestos")
    
elif 50000 >= ingr >= 10001:
    exc = ingr - 10000
    
    if edad>=65:
        redu = exc * 0.05
        exc1 = exc - redu
        result = exc1 - (1000*dep)
        print(f"Su total es {int(result*0.1)}")
        
    else:
        print(f"Su total es {int(exc*0.1)}")
    

elif 50000 < ingr:
    gra = ingr - (1000*dep)
    exc = gra - 50000
    exc1 = exc + 4000
    
    if edad > 65:
        redu = exc1 * 0.05
        impu = int(exc1*0.2)
        impuesto = impu-redu
        print(f"Su total es {int(impuesto)}")
    
    else:
        impuesto = exc1 - (1000-dep)
        print(f"Su total a pagar es {int(impuesto*0.2)}")
        
```
